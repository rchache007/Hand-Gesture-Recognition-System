```python
#Import required libraries
import numpy as num_p
import cv2
import mediapipe as media_p
import tensorflow as t_f
from tensorflow.keras.models import load_model
```

```python
#Creating MediaPipe object and configuring it
hand_tracking_object = media_p.solutions.hands
range_hands = hand_tracking_object.Hands(max_num_hands=1, min_detection_confidence=0.8)
media_pipe_utilities = media_p.solutions.drawing_utils
```

```python
hand_gesture_model = load_model('hand gesture model')
#loading the class names
file = open('gestures.names', 'r')
#loading the labels in the file by splitting them with " "
labels = file.read().split('\n')
file.close()
print("Names of classes:", labels)
```

```python
#Starting the video capturing
frames = cv2.VideoCapture(0)

while not False:
    #loading each frame
    _, frame = frames.read()
    i, j, k = frame.shape
    frame = cv2.flip(frame, 1)
    
    # convert the BGR format to RGB format
    window_to_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    # detecting hand landmarks
    output = range_hands.process(window_to_rgb)
    recognized_gesture = ''

    # This if loop will check if hand landmarks are detected and will only predict of hand landmarks are predicted
    if output.multi_hand_landmarks:
        lmarks = []
        #For loop will iterate over landmarks of each hand if there are multiple hand
        for landmark_set in output.multi_hand_landmarks:
            #Looping over each landmark in set of landmarks of each hand, processing it and storing it 'lmarks'
            for landmark in landmark_set.landmark:
                landmark_x = int(landmark.x * i)
                landmark_y = int(landmark.y * j)
                lmarks.append([landmark_x, landmark_y])
                
            #draw landmarks in the current frame
            media_pipe_utilities.draw_landmarks(frame, landmark_set, hand_tracking_object.HAND_CONNECTIONS)
            #making the predictions
            predictions = hand_gesture_model.predict([lmarks])
            #Getting the prediction with highest confidence
            cid = num_p.argmax(predictions)
            #getting the name of predicted gesture
            recognized_gesture = labels[cid]
            

    cv2.putText(frame, recognized_gesture, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 
                   1, (0,0,255), 2, cv2.LINE_AA)
    cv2.imshow("Output", frame) 

    if cv2.waitKey(1) == 27:
        break

frames.release()
cv2.destroyAllWindows()

```
